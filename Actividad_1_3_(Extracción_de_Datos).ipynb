## 1. Importar librerias

```python
import pandas as pd
```

## 2. Carga de archivo XLSX


```python
df = pd.read_excel('./Recursos/Inmuebles.xlsx', header=1)
display(df)
```

# 3. Filtros

### 3.1 Eliminar la columna “Referencia”

```python
filtro1 = df.drop(columns=['Referencia'], axis=1)

display(filtro1)

# Guardar filtro en CSV
filtro1.to_csv("./Filtros/Act1.3/Filtro1.csv")
```

### 3.2 Todos las registros de “Fecha Alta : igual a 2007-04-29 00:00:00 y a 2004-01-02 00:00:00

```python
filtro2 = df.loc[(df['Fecha_Alta'] == '2007-04-29 00:00:00') | (df['Fecha_Alta'] == '2004-01-02 00:00:00')]

# Imprimir los registros seleccionados
display(filtro2)

# Guardar filtro en CSV
filtro2.to_csv("./Filtros/Act1.3/Filtro2.csv")
```

### 3.3 Todos los registros de “ Tipo: parking, local, industrial y piso”

```python
filtro3 = df[df['Tipo'].isin(['Parking', 'Local', 'Industrial', 'Piso'])]
display(filtro3)

# Guardar filtro en CSV
filtro3.to_csv("./Filtros/Act1.3/Filtro3.csv")
```

### 3.4 Todos los registros con “Provincia” igual a “Girona” y con “Superficie” mayor a “100”

```python
filtro4 = df[(df['Provincia'] == 'Girona') & (df['Superficie'] > 100)]
display(filtro4)

# Guardar filtro en CSV
filtro4.to_csv("./Filtros/Act1.3/Filtro4.csv")
```

### 3.5 Todos los registros con “Precio Venta: mayor a 1000,000” o “vendedor” igual a “Carmen o Joaquin”

```python
filtro5 = df[(df['Precio Venta'] > 1000000) | (df['Vendedor'].isin(['Carmen', 'Joaquín']))]
display(filtro5)

# Guardar filtro en CSV
filtro5.to_csv("./Filtros/Act1.3/Filtro5.csv")
```

### 3.6 Todos los registros que no sean de “Provincia : Barcelona“

```python
filtro6 = df[df['Provincia'] != 'Barcelona']
display(filtro6)

# Guardar filtro en CSV
filtro6.to_csv("./Filtros/Act1.3/Filtro6.csv")
```

### 3.7 Todos los registros que no sean de “Operación: alquiler “ ni “Tipo: oficina“

```python
filtro7 = df[~((df['Operación'] == 'Alquiler') | (df['Tipo'] == 'Oficina'))]
display(filtro7)

# Guardar filtro en CSV
filtro7.to_csv("./Filtros/Act1.3/Filtro7.csv")
```
