# Manejo de valores atipicos pera tratando el DF completo

```python
import pandas as pd
import matplotlib.pyplot as plt
```

- 1. Leer archivo

```python
data = pd.read_csv('../Recursos/Ventas_totales.csv')

# Verificamos info del DF
data.info()


# Eliminar columnas innecesarias 
# axis=1 para col
# axis=0 para row
# Siendo que para trabajar con valores atipicos necesitamos datos numericos, la col indice_tiempo es irrelevante al tratarse de datos tipo objeto

data2 = data.drop(['indice_tiempo'], axis=1)
display(data2)

# Corroboramos valores nulos
valores_nulos = data.isnull().sum()
print(valores_nulos)
```

- 2. Procedimiento para eliminar valores nulos en DF

```python
data2 = data2.fillna(method='bfill')
display(data2)

# Corroboramos valores nulos
valores_nulos = data2.isnull().sum()
print(valores_nulos)
```

- 3. Realizamos diagrama de bigote a nivel DF

```python
fig = plt.figure(figsize = (15, 8))
data2.plot(kind='box', vert=False)
plt.title('Valores Atipicos del DF')
plt.show()
```

## 1. Metodo desviacion estandar

```python
# Encontramos los valores extremos
y = data2
limiteSuperior = y.mean() + 3*y.std()
limiteInferior = y.mean() - 3*y.std()

print(f"Limite superior permitido :")
for col, val in limiteSuperior.items():
    print(f"{col}: {round(val, 2)}")

print(f"\n\nLimite inferior permitido :")
for col, val in limiteInferior.items():
    print(f"{col}: {round(val, 2)}")
```

```python
# Encontramos Outliers
outliers = data2[(y>limiteSuperior) | (y<limiteInferior)]
print(f"\nOutliers:\n")
for col, val in outliers.items():
    print(f"{col}: {round(val, 2)}")
```

- 1.1 Datos limpios

```python
data3 = data2[(y<=limiteSuperior) & (y>=limiteInferior)] #Se convierten los outliers en nulos
display(data3)

# Reemplazamos los outliers con el mean de cada col
data_clean = data3.copy()
data_clean = data_clean.fillna(round(data3.mean(), 1))
display(data_clean)
```

```python
# Corroboramos valores nulos
valores_nulos = data_clean.isnull().sum()
print(valores_nulos)
```

## 2. Metodo cuantiles

```python
# Encontramos los valores extremos
y = data2

percentile25 = y.quantile(0.25) #Q1
percentile75 = y.quantile(0.75) #Q3

iqr = percentile75 - percentile25

limiteSuperiorIQR = percentile75 + 1.5 * iqr
limiteInferiorIQR = percentile25 - 1.5 * iqr


print(f"Limite superior permitido :")
for col, val in limiteSuperiorIQR.items():
    print(f"{col}: {round(val, 2)}")

print(f"\n\nLimite inferior permitido :")
for col, val in limiteInferiorIQR.items():
    print(f"{col}: {round(val, 2)}")
```

```python
# Encontramos Outliers
outliers = data2[(y>limiteSuperiorIQR) | (y<limiteInferiorIQR)]
print(f"\nOutliers:\n")
for col, val in outliers.items():
    print(f"{col}: {round(val, 2)}")
```

- 2.1 Convertir Outliers en nulos

```python
data3_iqr = data2[(y<=limiteSuperiorIQR) & (y>=limiteInferiorIQR)]
display(data3_iqr)
```

```python
# Corroboramos valores nulos
valores_nulos = data3_iqr.isnull().sum()
print(valores_nulos)
```

```python
# Reemplazamos los outliers con el mean de cada col
data4_iqr = data3_iqr.copy()
data4_iqr = data_clean.fillna(round(data3_iqr.mean(), 1))
display(data4_iqr)
```

```python
# Corroboramos valores nulos
valores_nulos = data4_iqr.isnull().sum()
print(valores_nulos)
```

```python
# Finalmente unimos la columna 'indice_tiempo' con el DF
indice_tiempo = data['indice_tiempo']
datos_limpios = pd.concat([indice_tiempo, data4_iqr], axis=1)
display(datos_limpios)
```
