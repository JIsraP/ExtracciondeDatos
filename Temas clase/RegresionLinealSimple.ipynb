# Regresion lineal simple

## 1. Importar Librerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Cargar archivo

```python
df = sns.load_dataset("car_crashes")
df.head()
```

## 3. Manejo del archivo

- 3.1 Mostrar correlacion de los datos

```python
sns.scatterplot(x="alcohol", y="total", color="blue", data=df)
sns.scatterplot(x="speeding", y="total", color="red", data=df)
```

```python
# Declaramos las variables dependientes e independientes para la regresion lineal
#varIndep = df[["alcohol", "speeding"]] #Se pueden almacenar varias variables a la vez

varsIndep = df[['alcohol']]
varDep = df['total']
```

## 4. Modelo regresion lineal

- 4.1 Uso de LinearRegression

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
type(model)
```

- 4.2 Ajustamos el modelo con las variables

```python
model.fit(X = varsIndep, y=varDep)
```

- 4.3 Obtenemos diccionario del modelo

```python
model.__dict__
```

- 4.4 Modelo matematico, y = 2.0325063x + 5.857776154826299 

```python
# Predecimos los valores de total de accidentes a partir de la variable "alcohol"
# yPred = model.predict(X = df[["alcohol", "speeding"]])

yPred = model.predict(X = df[["alcohol"]])
print(yPred)
```

- 4.5 Insertar prediccion al DF inicial

```python
df.insert(0, "alcohol_predict", yPred)
display(df)
```

- 4.6 Visualizamos grafica comparativa entre el total real y el total predecido

```python
sns.scatterplot(x='alcohol', y='total', color='blue', data=df)
sns.scatterplot(x='alcohol', y='alcohol_predict', color='red', data=df)
sns.lineplot(x='alcohol', y='alcohol_predict', color='red', data=df)
```

- 4.7 Coeficiente de determinacion r^2

```python
coefDeter = model.score(X=varsIndep, y=varDep)
print(coefDeter)
```

- 4.8 Coeficiente de correlacion r

```python
coefCorr = np.sqrt(coefDeter)
print(coefCorr)
```

- 4.9 Elimina la columnas tipo str

```python
df = df.drop(['abbrev'], axis=1)
display(df.head())
```
