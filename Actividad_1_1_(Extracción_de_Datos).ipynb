## 1. Importar librerias

```python
import pandas as pd
```

## 2. Carga de archivo XLSX


```python
df = pd.read_excel('./Recursos/Detalle precios y productos fabricados 2022.xlsx')
display(df)
```

# 3. Filtros

### 3.1 Primer filtro

```python
#Se filtran todos aquellos registros donde el nombre del vendedor sea LETICIA RAMIREZ HERNANDEZ
filtro1 = df[df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ"]
display(filtro1)

# Guardar filtro en CSV
filtro1.to_csv("./Filtros/Act1.1/Filtro1.csv")
```

### 3.2 Segundo filtro

```python
#Se filtra por filas, en este caso la fila con el indice 0, 50 y 100
filtro2 = df.iloc[[0,50,100], :]
display(filtro2)

# Guardar filtro en CSV
filtro2.to_csv("./Filtros/Act1.1/Filtro2.csv")
```

### 3.3 Tercer filtro

```python
#Filtro por columnas, de la 1 a la 5
filtro3 = df.iloc[:,1:5]
display(filtro3)

# Guardar filtro en CSV
filtro3.to_csv("./Filtros/Act1.1/Filtro3.csv")
```

### 3.4 Cuarto filtro

```python
#Volvemos a cargar el archivo xlsx para guardarlo en un nuevo DF, pero tomamos en cuenta la primer columna(CVE_DOC) como indice
df1 = pd.read_excel("./Recursos/Detalle precios y productos fabricados 2022.xlsx", index_col=0)

#Filtramos el nombre del vendedor y cliente por el indice indicado
filtro4 = df1.loc[["F39834", "F39837"],["NOMBRE_VENDEDOR", "NOMBRE_CLIENTE"]]
display(filtro4)

# Guardar filtro en CSV
filtro4.to_csv("./Filtros/Act1.1/Filtro4.csv")
```

### 3.5 Quinto filtro

```python
#Filtrar por medio de la funcion head, que toma los primeros registros. En este caso los primero 10
filtro5 = df.head(10)
display(filtro5)

# Guardar filtro en CSV
filtro5.to_csv("./Filtros/Act1.1/Filtro5.csv")
```

### Sexto filtro

```python
# Filtro por variables cuantitativas, en este caso la columna de PRECIO_UNITARIO en donde su valor sea mayor a 1000
filtro6 = df[df["PRECIO_UNITARIO"] > 1000]
display(filtro6)

# Guardar filtro en CSV
filtro6.to_csv("./Filtros/Act1.1/Filtro6.csv")
```

### Septimo filtro

```python
# Filtro con conector logico and, en este caso utilzamos la columna COSTO_TOTAL_CALCULADO y buscamos los valores que sean mayores a 10000, asimismo, utilizamos la columna NOMBRE_VENDEDOR y buscamos la coincidencia con el nombre LETICIA RAMIREZ HERNANDEZ. Ambos parametros deben de cumplirse para guardar el registro
filtro7 = df[(df["COSTO_TOTAL_CALCULADO"] > 10000) & (df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ")]
display(filtro7)

# Guardar filtro en CSV
filtro7.to_csv("./Filtros/Act1.1/Filtro7.csv")
```

### Octavo filtro

```python
# Filtro con conector logico or, en este caso utilzamos la columna COSTO_TOTAL_CALCULADO y buscamos los valores que sean menores a 100, asimismo, utilizamos la columna FECHA_DOC y buscamos la coincidencia con la fecha 2022-12-23. Con uno de los parametros que se cumpla el registro se guardarda
filtro8 = df[(df["COSTO_TOTAL_CALCULADO"] < 100) | (df["FECHA_DOC"] == "2022-12-23")]
display(filtro8)

# Guardar filtro en CSV
filtro8.to_csv("./Filtros/Act1.1/Filtro8.csv")
```

### Noveno filtro

```python
#Filtro con conector logico not, en este caso utilzamos la columna COSTO_TOTAL_CALCULADO y buscamos los valores que no sean menores a 100, asimismo, utilizamos la columna FECHA_DOC y buscamos que no coincidencia con la fecha 2022-12-23. Ambos parametros deben de cumplirse para guardar el registro
filtro9 = df[~(df["COSTO_TOTAL_CALCULADO"] < 100) & ~(df["FECHA_DOC"] == "2022-12-23")]
display(filtro9)

# Guardar filtro en CSV
filtro9.to_csv("./Filtros/Act1.1/Filtro9.csv")
```
