## 1. Importar librerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 2. Cargar archivo

```python
data2020 = pd.read_excel('./Recursos/Gastos y costos 20-23.xlsx', sheet_name='2020', header=5)
data2021 = pd.read_excel('./Recursos/Gastos y costos 20-23.xlsx', sheet_name='2021', header=3)
data2022 = pd.read_excel('./Recursos/Gastos y costos 20-23.xlsx', sheet_name='2022', header=3)
data2023 = pd.read_excel('./Recursos/Gastos y costos 20-23.xlsx', sheet_name='2023', header=3)

print("Datos 2020: ")
display(data2020)

print("\n\nDatos 2021: ")
display(data2021)

print("\n\nDatos 2022: ")
display(data2022)

print("\n\nDatos 2023: ")
display(data2023)
```

## 3. Datos 2020

- 3.1 Verificamos info del DF

```python
data2020.info()
```

```python
# Separamos columnas cuantitativas y cualitativas
data2020Cualitativas  = data2020.iloc[: , [0, 1, 2, 3, 4, 5, 6, 7, 15, 16, 17]]
data2020Cuantitativas = data2020.iloc[: , [8, 9, 10, 11, 12, 13, 14, 18]]


print(f'Columnas cualitativas: {data2020Cualitativas.columns}')
print(f'\n\nColumnas cuantitativas: {data2020Cuantitativas.columns}')
```

- 3.2 Verificar que no hayan valores nulos

```python
valores_nulos = data2020Cuantitativas.isnull().sum()
print(valores_nulos)
```

- 3.3 Eliminar valores nulos en el DF

```python
data2020Cuantitativas = data2020Cuantitativas.fillna(method='bfill')
display(data2020Cuantitativas)
```

```python
# Corroboramos valores nulos
valores_nulos = data2020Cuantitativas.isnull().sum()
print(valores_nulos)
```

- 3.4 Diagrama de Bigote

```python
fig = plt.figure(figsize = (15, 8))
data2020Cuantitativas.plot(kind='box', vert=False)
plt.title('Valores Atipicos del DF2020')
plt.show()
```

- 3.5 Metodo Desviacion Estandar

```python
# Encontramos los valores extremos
y = data2020Cuantitativas
limiteSuperior = y.mean() + 3*y.std()
limiteInferior = y.mean() - 3*y.std()

print(f"Limite superior permitido :")
for col, val in limiteSuperior.items():
    print(f"{col}: {round(val, 2)}")

print(f"\n\nLimite inferior permitido :")
for col, val in limiteInferior.items():
    print(f"{col}: {round(val, 2)}")
```

```python
# Encontramos Outliers
outliers = data2020Cuantitativas[(y>limiteSuperior) | (y<limiteInferior)]
print(f"\nOutliers:\n")
for col, val in outliers.items():
    print(f"{col}: {round(val, 2)}")
```

- 3.6 Generamos Datos Limpios

```python
data3 = data2020Cuantitativas[(y<=limiteSuperior) & (y>=limiteInferior)] #Se convierten los outliers en nulos
display(data3)
```

```python
# Reemplazamos los outliers con el mean de cada col
data_clean = data3.copy()
data_clean = data_clean.fillna(round(data3.mean(), 2))
display(data_clean)
```

```python
# Corroboramos valores nulos
valores_nulos = data_clean.isnull().sum()
print(valores_nulos)
```

```python
# Graficamos la grafica de boxplot, encontramos nuevos outliers, pero son distintos a los originales
fig = plt.figure(figsize = (15, 8))
data_clean.plot(kind='box', vert=False)
plt.title('Valores Atipicos del DF2020')
plt.show()
```

- 3.7 Unimos las columnas cuantitativas con las cualitativas

```python
datos_limpios = pd.concat([data2020Cualitativas, data_clean], axis=1)
display(datos_limpios)
```

- 3.8 Descargar DF

```python
datos_limpios.to_csv("./Filtros/Act3.2/datos2020.csv")
```
