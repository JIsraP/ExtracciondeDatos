## 1. Importar librerias

```python
import pandas as pd
```

## 2. Carga de archivo XLSX


```python
df = pd.read_excel('./Recursos/Datos de Facturación.xlsx')
display(df)
```

# 3. Filtros

### 3.1 CVE_CLPV (valores: 1000 a 2000)

```python
# Limpieza de datos con replace
df.CVE_CLPV = df.CVE_CLPV.replace({'O':'0'}, regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'6781|':'6781'}, regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'3612C':'3612'}, regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'      ':'0'}, regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'2385A':'2385'}, regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'2510A':'2510'}, regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'8440Q':'8440'}, regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'M':'0'}, regex=False)

# Convierte la columna 'CVE_CLPV' a enteros
df['CVE_CLPV'] = df['CVE_CLPV'].astype(float)

# Aplicar Filtro
filtro1 = df[(df['CVE_CLPV'] >= 1000) & (df['CVE_CLPV'] <= 2000)]
display(filtro1)

# Guardar filtro en CSV
filtro1.to_csv("./Filtros/Act1.2/Filtro1.csv")
```

### 3.2 CVE_VEND (Todas las claves excepto “5” y “4”)

```python
display(df.groupby(['CVE_VEND'])['CVE_VEND'].count())

# Filtrar las filas donde CVE_VEND no sea "4" ni "5"
filtro2 = df[~df['CVE_VEND'].isin([4, 5])]

display(filtro2.groupby(['CVE_VEND'])['CVE_VEND'].count())

# Guardar filtro en CSV
filtro2.to_csv("./Filtros/Act1.2/Filtro2.csv")
```

### 3.3 FECHA_ENT (Fechas de “13/04/2022  0:00:00”)

```python
df['FECHA_ENT'] = pd.to_datetime(df['FECHA_ENT'], format='%Y-%m-%d %H:%M:%S',  errors='coerce')

# Filtrar las filas donde FECHA_ENT sea igual a "13/04/2022 0:00:00"
fecha = pd.to_datetime("13/04/2022 0:00:00")
filtro3 = df[df['FECHA_ENT'] == fecha]

display(filtro3)

# Guardar filtro en CSV
filtro3.to_csv("./Filtros/Act1.2/Filtro3.csv")
```

### 3.4 CAN_TOT (Cantidades menores a 5951.7) o STATUS (“E”)

```python
filtro4 = df[(df['CAN_TOT'] < 5951.7) | (df['STATUS'] == 'E')]

# Mostrar el DataFrame filtrado
display(filtro4)

# Guardar filtro en CSV
filtro4.to_csv("./Filtros/Act1.2/Filtro4.csv")
```

### 3.5 Solo las columnas: CVE_DOC, FECHA_ENT, FECHA_VEN y CAN_TOT

```python
# Seleccionar las columnas deseadas
filtro5 = df[['CVE_DOC', 'FECHA_ENT', 'FECHA_VEN', 'CAN_TOT']]

display(filtro5)

# Guardar filtro en CSV
filtro5.to_csv("./Filtros/Act1.2/Filtro5.csv")
```

### 3.6 Solo las filas de 7001-7099

```python
filtro6 = df.iloc[7000:7100]
display(filtro6)

# Guardar filtro en CSV
filtro6.to_csv("./Filtros/Act1.2/Filtro6.csv")
```

### 3.7 Index= CVE_VEND (valores: 1, 2)  (columna: FECHAELAB)

```python
df.set_index("CVE_VEND", inplace=True)

filtro7 = df.loc[df.index.isin([1, 2]), "FECHAELAB"]
display(filtro7)

# Guardar filtro en CSV
filtro7.to_csv("./Filtros/Act1.2/Filtro7.csv")
```
